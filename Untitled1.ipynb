```python
def time(h,k,A,t):
    wcr = (2*3.1415)/t
    a = (4*h*K)/(3.1415*A)
    b = a**2 - 1
    T = (b**0.5)/wcr
    return T
```

```python
D = []
for i in range(len(Aa)):
    
    wcr = (2*3.1415)/Ti[i]

    arr = [wcr*Time[i]]
    d = np.arctan(arr)
    D.append((3.1415 - d[0])/wcr)
```
