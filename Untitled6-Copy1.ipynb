```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from yellowbrick.regressor import residuals_plot
from yellowbrick.regressor import prediction_error
```

```python
from numpy import asarray
```

```python
# Analysis Without Noise
```

```python
data = pd.read_csv('2ndorderfinal1.csv')
```

```python
data.tail(20)
```

```python
y1 = data['Time Constant(T1)']
```

```python
y2 = data['Time Constant(T2)']
```

```python
data.drop(['Time Constant(T1)'], axis = 1, inplace = True)
```

```python
data.drop(['Time Constant(T2)'], axis = 1, inplace = True)
```

```python
data
```

```python
X = data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size = 0.2, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor1 = DecisionTreeRegressor(random_state = 2)
regressor1.fit(X_train, y_train)
```

```python
 regressor1.score(X_test,y_test)
```

```python
from yellowbrick.regressor import PredictionError
```

```python
visualizer = PredictionError(regressor1,xlabel = '$y_pred$',ylabel='$y_test$')

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show()
```

```python
print(prediction_error(regressor1, X_train, y_train, X_test, y_test))
```

```python
print("\nResiduals Plot")
print(residuals_plot(regressor1, X_train, y_train, X_test, y_test))
```

```python
ym = X_test['Measured T1']
y_pred = regressor1.predict(X_test)
ym = ym.to_numpy()
y_test = y_test.to_numpy()
```
