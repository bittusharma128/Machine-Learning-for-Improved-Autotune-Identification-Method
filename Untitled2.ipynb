```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Analysis Without Noise
```

```python
data = pd.read_csv('projectfilefinal.csv')
```

```python
data
```

```python
y1 = data['Time Constant']
```

```python
y2 = data['Delay']
```

```python
data.drop(['Time Constant'], axis = 1, inplace = True)
```

```python
data.drop(['Delay'], axis = 1, inplace = True)
```

```python
data.drop(['Noisy A'], axis = 1, inplace = True)
```

```python
data.drop(['Estimated T(N)'], axis = 1, inplace = True)
```

```python
data.drop(['Estimated D(N)'], axis = 1, inplace = True)
```

```python
data
```

```python
X = data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size = 0.2, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor1 = DecisionTreeRegressor(random_state = 2)
regressor1.fit(X_train, y_train)
```

```python
y_pred = regressor1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
 regressor1.score(X_train,y_train)
```

```python
from yellowbrick.regressor import residuals_plot
from yellowbrick.regressor import prediction_error
```

```python
from yellowbrick.regressor import PredictionError
```

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
visualizer = PredictionError(regressor1,xlabel = '$y_pred$',ylabel='$y_test$')

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show()  
```

```python
visualizer = ResidualsPlot(
    regressor1,
    hist=False, 
    qqplot=True,
    train_color="gold",
    test_color="maroon")
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

```python
print(prediction_error(regressor1, X_train, y_train, X_test, y_test))
```

```python
print("\nResiduals Plot")
print(residuals_plot(regressor1, X_train, y_train, X_test, y_test))
```
